```python
df=pd.read_excel('drive/My Drive/fp_421/2021-2022 NBA Player Stats.xlsx', index_col=0)
df.columns = ['Player_name', 'team', 'pos', 'age', 'gp', 'mpg', 'min%', 'usg%', 'tor%', 'fta', 'ft%', '2pa', '2p%', '3pa', '3p%', 'efg%', 'ts%', 'ppg', 'rpg', 'trb%', 'apg', 'ast%', 'spg', 'bpg', 'topg', 'vi', 'ortg', 'drtg']
df = df.iloc[1: , :]
df = df.reset_index()
df.drop(['2021-2022 NBA Player Stats'], axis = 1, inplace=True)
df['year'] = '2021-2022'
teams= ['Lac', 'Pho', 'Atl', 'Mil']
df['Final_four'] = df['team'].isin(teams).astype(int)
df.head()
```

```python
fnames = ['2018-2019 NBA Player Stats', '2019-2020 NBA Player Stats', '2020-2021 NBA Player Stats']
final_teams = [['Hou', 'Gol', 'Cle', 'Bos'],['Gol', 'Ind', 'Tor', 'Mil'], ['Lal', 'Den', 'Mia', 'Bos']]
```

```python
i=0
for fname in fnames:
    name = fname+'.xlsx'
    year = fname[0:9]
    data = pd.read_excel('drive/My Drive/fp_421/'+name, index_col = 0)
    data.columns = ['Player_name', 'team', 'pos', 'age', 'gp', 'mpg', 'min%', 'usg%', 'tor%', 'fta', 'ft%', '2pa', '2p%', '3pa', '3p%', 'efg%', 'ts%', 'ppg', 'rpg', 'trb%', 'apg', 'ast%', 'spg', 'bpg', 'topg', 'vi', 'ortg', 'drtg']
    data['year']=year
    data = data.iloc[1: , :]
    data = data.reset_index()
    data.drop(fnames[i], axis = 1, inplace=True)
    data['Final_four'] = data['team'].isin(final_teams[i]).astype(int)
    df = df.append(data)
    i+=1
```
